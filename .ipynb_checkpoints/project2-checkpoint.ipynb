```python
"""
Template for week 3 project in Data Visualization

Read, clean, and plot financial time series from stock indices using matplotlib
"""

import csv
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Resource paths
DATA_PATH = "data/"
PLOTS_PATH = "plots/"

# Financial data files
DJIA_34 = DATA_PATH + "DJIA_34.csv"
DJIA = DATA_PATH + "DJIA.csv"
NASDAQ_100 = DATA_PATH + "NASDAQ_100.csv"
NASDAQ = DATA_PATH + "NASDAQ.csv"
SP500 = DATA_PATH + "SP500.csv"
```

```python
###################################################################
# Part 1 - Read, clean, and plot financial time series data 

def read_series(index_file):
    """
    Input: String index_file for CSV file containing stock index data
    
    Output: Time series as a list of tuples where each tuple is
    a pair of strings corresponding to a row in the CSV file
    """ 
    data = []
    with open(index_file, newline = '') as csv_file:
        csv_reader = csv.reader(csv_file)
        for row in csv_reader:
            if len(row) == 2:
                data.append((row[0], row[1]))
    
    return data
```

```python
def clean_series(raw_series):
    """
    Input: List raw_series whose entries are tuples (date, price)
    where dates corresponding to holidays have price "."
    
    Output: A list whose entries are tuples consisting of
    a date (string) and a price (float). Entries corresponding
    to the header and holidays have been removed.
    """
    clean_data = []
    
    #Remove header tuple
    no_header_series = raw_series[1:]
    
    #Remove tuples corresponding to weekday holidays
    for index in range(len(no_header_series) - 1, -1, -1):
        dummy_date, dummy_price = no_header_series[index]
        if dummy_price == '.':
            no_header_series.pop(index)
    
    #Convert remaining price data to floats
    for dummy_date, dummy_price in no_header_series:
        clean_data.append((dummy_date, float(dummy_price)))

    return clean_data
```

```python
def plot_prices(index_file):
    """
    Input: String index_file corresponding to a csv file
    containing financial time series data
    
    Output: matplotlib figure consistin of a line plot 
    of prices from cleaned series
    y-axis should have appropriate label
    """
    #Read and clean file
    cleaned_data = clean_series(read_series(index_file))
    
    date_list = []
    float_price = []
    for dummy_date, dummy_price in cleaned_data:
        date_list.append(dummy_date)
        float_price.append(dummy_price)
    
    fig, axs = plt.subplots()
    axs.plot(date_list, float_price)
    axs.set_ylabel("Index Price (Feb 5, 1971=100)")
    
    return fig
```

```python
def test_plot_prices():
    """ Test plot_prices """
    
    plot_prices(DJIA_34)
    plot_prices(NASDAQ_100)

test_plot_prices()
```

```python
###########################################################
# Part 2 - Convert ISO data format to days and make better plots

def isodate_to_day(isodate):
    """
    Input: A string date in ISO date format

    Output: Float corrspending to mumber of days used in
    matplotlib's plt.plot_date()
    """
    #Convert string to datetime object
    dt_object = datetime.datetime.strptime(isodate, "%Y-%m-%d")
    
    #Convert datetime object to matplotlib date (float) with return
    return mdates.date2num(dt_object)
```

```python
###########################################################
# Student code for more plots in matplotlib (peer-graded)

def plot_dates_prices(index_file):
    """
    Input: Strings index_file for CSV file with stock index data
    
    Output: matplotlib figure generated by reading the CSV file
    index_file and plotting the dates/prices using isodate_to_day()
    """
    #Clean and read index file
    cleaned_data = clean_series(read_series(index_file))
    
    #Create dates and price (floats) list for plotting
    dates_list = []
    floats_price = []
    for dummy_date, dummy_price in cleaned_data:
        dates_list.append(isodate_to_day(dummy_date))
        floats_price.append(dummy_price)
    
    fig, axs = plt.subplots(figsize=(20,15))
    axs.plot_date(dates_list, floats_price, linestyle='-', alpha=0.8,
                  linewidth=0.01)
    axs.set_xlabel("Year", fontsize=20)
    axs.set_ylabel("Index Price (Feb 5, 1971=100)", fontsize=20)
    axs.set_title(f"{index_file[5:-4]}", fontsize=20)
      
    return fig
```

```python
def test_plot_dates_prices():
    """ Test plot_dates_prices"""

    plot_dates_prices(DJIA)
    plot_dates_prices(NASDAQ)
    plot_dates_prices(SP500)
    
test_plot_dates_prices()
```

```python
def plot_two_indices(index_file1, index_file2):
    """
    Input: Strings index_file1 and index_file2
    with CSV data for pair of stock indices
    
    Output: matplotlib figure generated by plotting a comparison
    of the price of two indices in the style of 
    https://matplotlib.org/gallery/api/two_scales.html
    """
    #Read and clean the respective files
    clean_file1 = clean_series(read_series(index_file1))
    clean_file2 = clean_series(read_series(index_file2))
    
    #create date and price lists for plotting from cleaned lists (for each series)
    dates_list_1 = []
    floats_price_1 = []
    for dummy_date, dummy_price in clean_file1:
        dates_list_1.append(isodate_to_day(dummy_date))
        floats_price_1.append(dummy_price)
        
    dates_list_2 = []
    floats_price_2 = []
    for dummy_date, dummy_price in clean_file2:
        dates_list_2.append(isodate_to_day(dummy_date))
        floats_price_2.append(dummy_price)
    
    #Plots for each figure
    fig, ax1 = plt.subplots(figsize=(20,15))
    
    #Figure one formatting
    ax1.plot_date(dates_list_1, floats_price_1, color="tab:red", linestyle='-',
                  linewidth=0.01, alpha=0.8)
    ax1.set_xlabel("Year", fontsize=20)
    ax1.set_ylabel(f"Index Price for {index_file1[5:-4]} (Feb 5, 1971=100)",
                   color="tab:red", fontsize=20)
    
    #Figure 2
    ax2 = ax1.twinx()
    ax2.plot_date(dates_list_2, floats_price_2, color="tab:blue", linestyle='-',
                 alpha=0.8, linewidth=0.01)
    ax2.set_ylabel(f"Index Price for {index_file2[5:-4]} (Feb 5, 1971=100)",
                   color="tab:blue", fontsize=20)
    
    ax1.set_title(f"{index_file1[5:-4]} vs {index_file2[5:-4]} Index Prices",
                 fontsize=20)
    
    return fig
```

```python
def test_plot_two_indices():
    """ Test plot_two_indices """
    
    plot_two_indices(NASDAQ, SP500)
    plot_two_indices(NASDAQ, DJIA)
    
test_plot_two_indices()
```
